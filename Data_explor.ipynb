```python
headers = {'Authorization': 'Bearer APP_USR-8132553454451839-033010-7297ef4dfba6ec5cccd11eb7fb42876f-831918494'}
 
productos = []

# Crear una lista vacía para almacenar los datos de los items
items_data = []

url = 'https://api.mercadolibre.com/products/'

# Recorrer el DataFrame y hacer una consulta en la API para cada ID
for idx, row in df_concatenado.iterrows():

    id = row['id']

    type = row['type']
    
    if type == 'PRODUCT':
        print(id)
        
        # Hacer la consulta en la API
        response = requests.get(url + id, headers=headers)

        if response.status_code == 200:
            
            # Determinar el formato de la respuesta
            if response.headers['content-type'] == 'application/json':
                item = response.json()
            elif response.headers['content-type'] == 'application/x-ndjson':
                item = response.json_lines()

    
            # Extraer la información relevante de cada item
            item_id = item['id']
            name = item['name']
            status = item['status']
            sold_quantity = item['sold_quantity']
            domain_id = item['domain_id']
            permalink = item['permalink']
            if item['buy_box_winner'] is not None:
                buy_box_winner_price = item['buy_box_winner']['price']
                buy_box_winner_currency_id = item['buy_box_winner']['currency_id']
                category_id = item['buy_box_winner']['category_id']
                seller_id = item['buy_box_winner']['seller_id']
                seller_city = item['buy_box_winner']['seller_address']['city']['name']
                seller_state = item['buy_box_winner']['seller_address']['state']['name']
            else:
                buy_box_winner_price = None
                buy_box_winner_currency_id = None
                category_id = None
                seller_id = None
                seller_city = None
                seller_state = None
            picture_url = item['pictures'][0]['url']
            brand = next((attr['value_name'] for attr in item['attributes'] if attr['id'] == 'BRAND'), '')
            model = next((attr['value_name'] for attr in item['attributes'] if attr['id'] == 'MODEL'), '')
            color = next((attr['value_name'] for attr in item['attributes'] if attr['id'] == 'COLOR'), '')
            with_bluetooth = next((attr['meta']['value'] for attr in item['attributes'] if attr['id'] == 'WITH_BLUETOOTH'), False)
            with_usb = next((attr['meta']['value'] for attr in item['attributes'] if attr['id'] == 'WITH_USB'), False)
            includes_remote_control = next((attr['meta']['value'] for attr in item['attributes'] if attr['id'] == 'INCLUDES_REMOTE_CONTROL'), False)

            # Crear un diccionario con los datos del item
            item_dict = {
                'item_id': item_id,
                'name': name,
                'status': status,
                'sold_quantity': sold_quantity,
                'domain_id': domain_id,
                'permalink': permalink,
                'buy_box_winner_price': buy_box_winner_price,
                'buy_box_winner_currency_id': buy_box_winner_currency_id,
                'category_id': category_id,
                'seller_id': seller_id,
                'seller_city': seller_city,
                'seller_state': seller_state,
                'picture_url': picture_url,
                'brand': brand,
                'model': model,
                'color': color,
                'with_bluetooth': with_bluetooth,
                'with_usb': with_usb,
                'includes_remote_control': includes_remote_control
            }

            # Agregar el diccionario a la lista de items
            items_data.append(item_dict)
        

# Crear un dataframe a partir de la lista de items
items_df = pd.DataFrame(items_data)
```
